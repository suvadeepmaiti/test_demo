#Opencv reads the image as BGR format; normally we see in RGB format.

```python
import numpy as np
from matplotlib import pyplot as plt
import cv2
img1=cv2.imread("/content/drive/MyDrive/IIIT Hyderabad/Assignment/Images/02_27162354_c4a4b5_2660065a.jpg")
img1=cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)    #Convert BGR to RGB 
img2=cv2.imread("/content/drive/MyDrive/IIIT Hyderabad/Assignment/Images/304442C800000578-3403987-image-a-32_1453064019909.jpg")
img2=cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)    #Convert BGR to RGB 
img3=cv2.imread("/content/drive/MyDrive/IIIT Hyderabad/Assignment/Images/3152F19F00000578-0-image-m-39_1455753105115.jpg")
img3=cv2.cvtColor(img3,cv2.COLOR_BGR2RGB)    #Convert BGR to RGB 
img4=cv2.imread("/content/drive/MyDrive/IIIT Hyderabad/Assignment/Images/5684f6ab86def6b700000001.jpg")
img4=cv2.cvtColor(img4,cv2.COLOR_BGR2RGB)    #Convert BGR to RGB
fig=plt.figure(figsize=(10,10))
fig.add_subplot(2,2,1)
plt.imshow(img1)
plt.axis('off')
plt.title('first')
fig.add_subplot(2,2,2)
plt.imshow(img2)
plt.axis('off')
plt.title('second')
fig.add_subplot(2,2,3)
plt.imshow(img3)
plt.axis('off')
plt.title('Third')
fig.add_subplot(2,2,4)
plt.imshow(img4)
plt.axis('off')
plt.title('Fourth') 
```
